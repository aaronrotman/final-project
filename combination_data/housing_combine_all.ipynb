```python
redfin_frames = [midwest_combined_csv, westcoast_combined_csv, east_combined_csv, south_combined_csv]
```

```python
redfin_all = pd.concat(redfin_frames)
```

```python
redfin_all.head(2)
```

```python
redfin_drop = redfin_all.drop(columns=['Unnamed: 0'])
```

```python
redfin_drop
```

```python
redfin_all_data = redfin_drop.reset_index(drop=True)
```

```python
redfin_all_data.dtypes
```

```python

redfin_all_data.loc[:,'city'] = redfin_all_data.loc[:, 'city'].astype("string")
redfin_all_data.loc[:,'state'] = redfin_all_data.loc[:, 'state'].astype("string")

```

```python
redfin_all_data.dtypes
```

```python
redfin_all_data_sorted = redfin_all_data.sort_values(by=['state', 'city'])
redfin_all_data_sorted
```

```python
redfin_all_data_sorted.to_csv('../redfin_housing/redfin_resources/combined_regions/redfin_all_regions.csv')
```
