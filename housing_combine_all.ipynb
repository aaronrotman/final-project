```python
import pandas as pd
```

```python
midwest_combined_csv = pd.read_csv("redfin_housing/redfin_resources/combined_regions/redfin_midwest_combined.csv")
westcoast_combined_csv = pd.read_csv("redfin_housing/redfin_resources/combined_regions/redfin_westcoast_combined.csv")
east_combined_csv = pd.read_csv("redfin_housing/redfin_resources/combined_regions/redfin_east_coast_combined.csv")


```

```python
redfin_frames = [midwest_combined_csv, westcoast_combined_csv, east_combined_csv]
```

```python
redfin_all = pd.concat(redfin_frames)
```

```python
redfin_all.head(2)
```

```python
redfin_drop = redfin_all.drop(columns=['Unnamed: 0'])
```

```python
redfin_drop
```

```python
redfin_all_data = redfin_drop.reset_index(drop=True)
```

```python
redfin_all_data.to_csv('redfin_housing/redfin_resources/combined_regions/redfin_all_regions.csv')
```

```python
population = pd.read_csv('population/population_categories.csv')
```

```python
population.head(2)
```

```python
temp_data = pd.read_csv('temperature/temp_min_and_max.csv')
```

```python
temp_data.head(2)
```

```python
pop_temp_merge = population.merge(temp_data, left_on='state', right_on='state_full',
          suffixes=('_pop', '_temp'))
```

```python
pop_temp_merge.head(2)
```

```python
pop_temp_drop_col = pop_temp_merge.drop(columns=['state_full'])
```

```python
pop_temp_drop_col.head(2)
```

```python
pop_temp_rename = pop_temp_drop_col.rename(columns={"state_pop": "state", "state_temp": "state_abbr"})
```

```python
pop_temp_rename.head(2)
```

```python
redfin_all_data.head(2)
```

```python
redfin_sort_state = redfin_all_data.sort_values(by=['state', 'city'])
```

```python
redfin_sort_state
```

```python
redfin_dropna = redfin_sort_state.dropna()
```

```python
redfin_dropna
```

```python
pop_temp_rename.sort_values(by=['state_abbr', 'city'])
```

```python
housing_complete_merge = redfin_dropna.merge(pop_temp_rename, left_on=['state', 'city'], right_on=['state_abbr', 'city'])
```

```python
housing_complete_merge
```

```python
housing_complete_drop = housing_complete_merge.drop(columns=['state_y', 'state_x'])
```

```python
housing_complete_drop
```

```python
housing_complete_drop['house_id'] = housing_complete_drop.index + 1
```

```python
housing_complete_drop
```

```python
cols = list(housing_complete_drop.columns.values)
cols
```

```python
complete_housing_data = housing_complete_drop[['house_id','property_type','address',
                                                'city','state_abbr','zip_code','price','beds','baths',
                                                'square_feet','lot_size','year_built','hoa_permonth',
                                                'population','pop_cat','min_temp','max_temp',
                                                'latitude','longitude','url']]
```

```python
complete_housing_data
```

```python
complete_housing_data.to_csv('housing_complete_data.csv')
```

```python
housing_training_data = complete_housing_data[['house_id','price','beds','baths','square_feet','lot_size',
                                               'hoa_permonth','population','pop_cat','min_temp','max_temp']]
```
