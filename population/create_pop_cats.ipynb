```python
import pandas as pd
import io
```

```python
population_csv = pd.read_csv("population/population_sortby_state.csv")
```

```python
population_csv
```

```python
# Create a list to store the data
pop_cat = []

# For each row in the column,
for row in population_csv['population']:
    # if more than a value,
    if row > 50000:
        # Append a letter grade
        pop_cat.append('large')
    # else, if more than a value,
    elif row > 2500:
        # Append a letter grade
        pop_cat.append('middle')
    else:
        # Append a failing grade
        pop_cat.append('small')
        
# Create a column from the list
population_csv['pop_cat'] = pop_cat
```

```python
population_csv
```

```python
population_csv_drop = population_csv.drop(columns=['location'])
```
