```python
import pandas as pd
import numpy as np
```

```python
boston_ma = pd.read_csv("redfin_resources/redfin_boston_ma.csv")
miami_fl = pd.read_csv("redfin_resources/redfin_miami_fl.csv")
nashville_tn = pd.read_csv("redfin_resources/redfin_nashville_tn.csv")
newyork_ny = pd.read_csv("redfin_resources/redfin_newyork_ny.csv")
washington_dc = pd.read_csv("redfin_resources/redfin_washington_dc.csv")
```

```python
boston_ma.head()
```

```python
washington_dc.head()
```

```python
frames = [boston_ma, miami_fl, nashville_tn, newyork_ny, washington_dc]
redfin_east_coast_combine = pd.concat(frames, ignore_index=True)
```

```python
redfin_east_coast_combine
```

```python
redfin_east_coast = redfin_east_coast_combine.drop(columns=['Unnamed: 0'])
```

```python
redfin_east_coast
```

```python
redfin_east_coast['hoa_permonth'] = redfin_east_coast['hoa_permonth'].fillna(0)
```

```python
redfin_east_coast['lot_size'] = redfin_east_coast['lot_size'].fillna(0)
```

```python
redfin_east_coast['beds'] = redfin_east_coast['beds'].fillna(0)
```

```python
redfin_east_coast['baths'] = redfin_east_coast['baths'].fillna(0)
```

```python
redfin_east_coast['square_feet'] = redfin_east_coast['square_feet'].fillna(0)
```

```python
redfin_east_coast['year_built'] = redfin_east_coast['year_built'].fillna('unknown')
```

```python
redfin_east_coast_dropna = redfin_east_coast.dropna()
redfin_east_coast_dropna
```

```python
east_coast_new_index = redfin_east_coast_dropna.reset_index(drop=True)
```

```python
east_coast_new_index
```

```python
east_coast_new_index.to_csv('redfin_resources/combined_regions/redfin_eastcoast_combined.csv')
```
