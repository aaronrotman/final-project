```python
import pandas as pd
import numpy as np
```

```python
boise_id = pd.read_csv("redfin_resources/redfin_boise_id.csv")
losangeles_ca = pd.read_csv("redfin_resources/redfin_losangeles_ca.csv")
seattle_wa = pd.read_csv("redfin_resources/redfin_seattle_wa.csv")
portlan_or = pd.read_csv("redfin_resources/redfin_portlan_or.csv")
sanfrancisco_ca = pd.read_csv("redfin_resources/redfin_sanfrancisco_ca.csv")
lasvegas_nv = pd.read_csv("redfin_resources/redfin_lasvegas_nv.csv")
sandiego_ca = pd.read_csv("redfin_resources/redfin_sandiego_ca.csv")
alaska = pd.read_csv("redfin_resources/alaska_all.csv")
arizona = pd.read_csv("redfin_resources/arizona_all.csv")

```

```python
frames = [boise_id, losangeles_ca, sandiego_ca, sanfrancisco_ca, lasvegas_nv, portlan_or, seattle_wa, alaska, arizona]
redfin_west_coast_combine = pd.concat(frames, ignore_index=True)
```

```python
redfin_west_coast_combine
```

```python
redfin_west_coast = redfin_west_coast_combine.drop(columns=['Unnamed: 0'])
```

```python
redfin_west_coast
```

```python
redfin_west_coast['hoa_permonth'] = redfin_west_coast['hoa_permonth'].fillna(0)
```

```python
redfin_west_coast['lot_size'] = redfin_west_coast['lot_size'].fillna(0)
```

```python
redfin_west_coast['beds'] = redfin_west_coast['beds'].fillna(0)
```

```python
redfin_west_coast['baths'] = redfin_west_coast['baths'].fillna(0)
```

```python
redfin_west_coast['square_feet'] = redfin_west_coast['square_feet'].fillna(0)
```

```python
redfin_west_coast['year_built'] = redfin_west_coast['year_built'].fillna('unknown')
```

```python
redfin_west_coast_dropna = redfin_west_coast.dropna()
redfin_west_coast_dropna
```

```python
west_coast_new_index = redfin_west_coast_dropna.reset_index(drop=True)
```

```python
west_coast_new_index
```

```python
west_coast_new_index.to_csv('redfin_resources/combined_regions/redfin_westcoast_combined.csv')
```
