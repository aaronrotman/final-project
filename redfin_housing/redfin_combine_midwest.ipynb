```python
import pandas as pd
import numpy as np
```

```python
littleRock_ar = pd.read_csv("redfin_resources/redfin_littleRockAR.csv")
milwaukee_wi = pd.read_csv("redfin_resources/redfin_milwaukeeWI.csv")
nevada_mo = pd.read_csv("redfin_resources/redfin_nevadaMO.csv")
okcity_ok = pd.read_csv("redfin_resources/redfin_okcityOK.csv")
```

```python
frames = [milwaukee_wi, littleRock_ar, nevada_mo, okcity_ok]
redfin_midwest_combine = pd.concat(frames, ignore_index=True)
```

```python
redfin_midwest_combine
```

```python
redfin_midwest = redfin_midwest_combine.drop(columns=['Unnamed: 0'])
```

```python
redfin_midwest
```

```python
redfin_midwest['hoa_permonth'] = redfin_midwest['hoa_permonth'].fillna(0)
```

```python
redfin_midwest['lot_size'] = redfin_midwest['lot_size'].fillna(0)
```

```python
redfin_midwest['beds'] = redfin_midwest['beds'].fillna(0)
```

```python
redfin_midwest['baths'] = redfin_midwest['baths'].fillna(0)
```

```python
redfin_midwest['square_feet'] = redfin_midwest['square_feet'].fillna(0)
```

```python
redfin_midwest['year_built'] = redfin_midwest['year_built'].fillna('unknown')
```

```python
redfin_midwest_dropna = redfin_midwest.dropna()
redfin_midwest_dropna
```

```python
midwest_new_index = redfin_midwest_dropna.reset_index(drop=True)
```

```python
midwest_new_index
```

```python
midwest_new_index.to_csv('redfin_resources/combined_regions/redfin_midwest_combined.csv')
```
