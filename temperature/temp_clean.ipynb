```python
import pandas as pd
```

```python
tmax = pd.read_csv('original_files/tmax.csv')
tmin = pd.read_csv('original_files/tmin.csv')
```

```python
tmax.head(2)
```

```python
tmin.head(2)
```

```python
tmax_temp_only = tmax.drop(columns=['Location ID', 'Rank', 'Anomaly (1901-2000 base period)', '1901-2000 Mean'])
tmin_temp_only = tmin.drop(columns=['Location ID', 'Rank', 'Anomaly (1901-2000 base period)', '1901-2000 Mean'])
```

```python
tmax_temp_only.head(2)
```

```python
tmin_temp_only.head(2)
```

```python
tmax_rename = tmax_temp_only.rename(columns={"Location": "state", "Value": "max_temp"})
tmin_rename = tmin_temp_only.rename(columns={"Location": "state", "Value": "min_temp"})
```

```python
temp_joined = tmin_rename.join(tmax_rename, lsuffix='_caller', rsuffix='_other')
```

```python
combined_drop = temp_joined.drop(columns=['state_other'])
combined_drop
```

```python
combined_rename = combined_drop.rename(columns={"state_caller": "state"})
combined_rename
```

```python
state_abbr = {
"Alabama":"AL","Arizona":"AZ","Arkansas":"AR","California":"CA",
"Colorado":"CO","Connecticut":"CT","Delaware":"DE","Florida":"FL","Georgia":"GA",
"Idaho":"ID","Illinois":"IL","Indiana":"IN","Iowa":"IA","Kansas":"KS","Kentucky":"KY","Louisiana":"LA",
"Maine":"ME","Maryland":"MD","Massachusetts":"MA","Michigan":"MI","Minnesota":"MN",
"Mississippi":"MS","Missouri":"MO","Montana":"MT","Nebraska":"NE","Nevada":"NV",
"NewHampshire":"NH","NewJersey":"NJ","NewMexico":"NM","NewYork":"NY","NorthCarolina":"NC",
"NorthDakota":"ND","Ohio":"OH","Oklahoma":"OK","Oregon":"OR","Pennsylvania":"PA",
"RhodeIsland":"RI","SouthCarolina":"SC","SouthDakota":"SD","Tennessee":"TN",
"Texas":"TX","Utah":"UT","Vermont":"VT","Virginia":"VA","Washington":"WA",
"WestVirginia":"WV","Wisconsin":"WI","Wyoming":"WY"
}

```

```python
state_df = pd.DataFrame.from_dict(state_abbr, orient='index')
```

```python
state_index = state_df.reset_index()
```

```python
state_rename = state_index.rename(columns={"index": "state_full", 0:"state"})
state_rename
```

```python
state_rename
```

```python
temp_state_abbr = combined_rename.join(state_rename, lsuffix='_caller', rsuffix='_other')
```

```python
temp_state_abbr
```

```python
temp_combo_drop = temp_state_abbr.drop(columns=['state_full'])
temp_combo_drop
```

```python
temp_min_max = temp_combo_drop.rename(columns={"state_caller": "state_full", "state_other":"state"})
```

```python
temp_min_max
```

```python
temp_min_max['temp_range'] = temp_min_max['max_temp'] - temp_min_max['min_temp']
```

```python
temp_min_max
```

```python
temp_min_max['temp_range'].min()
```

```python
temp_min_max['temp_range'].max()
```
